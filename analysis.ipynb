```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("data_clean.csv")
```

```python
data["Month"] = data.Date.apply(lambda x: int(x.split("-")[1]))
```

```python
data.head()
```

```python
months = ["January", "February", "March", "April", "May", "June", "July", "August", "September", "October", "November", "December"]
```

```python
month_average_low = np.array([data[data.Month == month]["low(C)"].mean() for month in range(1, 13)])
month_average_high = np.array([data[data.Month == month]["high(C)"].mean() for month in range(1, 13)])
```

```python
plt.figure(figsize=(15, 5))

plt.subplot(1, 2, 1)
plt.title("Average lowest temperature (degrees Celcius)")
sns.barplot(x=months, y=month_average_low)
plt.xticks(rotation=45)

plt.subplot(1, 2, 2)
plt.title("Average highest temperature (degrees Celcius)")
sns.barplot(x=months, y=month_average_high)
plt.xticks(rotation=45)

plt.show()
```

```python
month_average_daylight = np.array([data[data.Month == month]["daylight_hours(hours)"].mean() for month in range(1, 13)])
```

```python
plt.figure(figsize=(15, 5))

plt.title("Average daylight hours")
sns.barplot(x=months, y=month_average_daylight)
plt.xticks(rotation=45)

plt.show()
```
